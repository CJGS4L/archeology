```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

path = '/content/Kurgan_tabl_fin_dlya_stati (1).xlsx'
df1 = pd.read_excel(path)
df1
```

```python
grouped = (df1.groupby('слой').size().reset_index(name='Counts').sort_values('Counts'))
plt.figure(figsize=(8, 5))
grouped.plot.barh(x='слой', y='Counts', legend=False)
plt.title('Распределение количества объектов по слоям')
plt.xlabel('Количество записей в табличке')
plt.show()
#Предлагаю склеить все в "Другое" начиная с "желтое пятно" и дальше
```

```python
df = pd.read_excel(path)
df
df['слой'] = np.where((df['слой'] == 'серый, пятно')
|( df['слой'] == 'желтое пятно')
|( df['слой'] == 'граница серо-черного песка и желтого песка')
|( df['слой'] == 'серое пятно')
|( df['слой'] == 'под дерном')
|( df['слой'] == 'корневая яма')
|( df['слой'] == 'граница пепельного слоя и серо-желтого песка')
|( df['слой'] == 'слой палево-желтого песка и граница серо-желтого')
|( df['слой'] == 'заполнение ямы')
|( df['слой'] == 'серо-желтый песок')
|( df['слой'] == 'слой серо-черного песка')
|( df['слой'] == 'серый песок')
|( df['слой'] == 'темное пятно')
|( df['слой'] == 'под склоном')
|( df['слой'] == 'дерн низ')
|( df['слой'] == 'бордовое пятно')
|( df['слой'] == 'Unknown') , 'другое', df['слой'])

df['слой'] = np.where((df['слой'] == 'серый, пятно')
|( df['слой'] == 'граница пепельного и серо-желтого песка'), 'граница пепельного песка и серо-желтого песка', df['слой'])

df['слой'] = np.where((df['слой'] == 'серый, пятно')
|( df['слой'] == 'слой палево-желтого песка'), 'палево-желтый песок', df['слой'])

df['слой'] = np.where((df['слой'] == 'серый, пятно')
|( df['слой'] == 'слой желтого песка'), 'желтый песок', df['слой'])
```

```python
grouped = (df.groupby('слой').size().reset_index(name='Counts').sort_values('Counts'))
plt.figure(figsize=(8, 5))
grouped.plot.barh(x='слой', y='Counts', legend=False)
plt.title('Распределение количества объектов по слоям')
plt.xlabel('Количество записей в табличке')
plt.show()
#Предлагаю склеить все в "Другое" начиная с "серый, пятно" и дальше
```

```python
grouped = (df1.groupby('НОВЫЙ ТИП СЫРЬЯ').size().reset_index(name='Counts').sort_values('Counts'))
plt.figure(figsize=(8, 5))
grouped.plot.barh(x='НОВЫЙ ТИП СЫРЬЯ', y='Counts', legend=False)
plt.title('Распределение количества объектов по атрибуту "Новый тип сырья"')
plt.xlabel('Количество записей в табличке')
plt.show()
```

```python
df['НОВЫЙ ТИП СЫРЬЯ'] = np.where((df['НОВЫЙ ТИП СЫРЬЯ'] == 'серый, пятно')
|( df['НОВЫЙ ТИП СЫРЬЯ'] == 'кварцит')
|( df['НОВЫЙ ТИП СЫРЬЯ'] == 'полупрозрачный бежевый кремень (специфический тип)')
|( df['НОВЫЙ ТИП СЫРЬЯ'] == 'известняк')
|( df['НОВЫЙ ТИП СЫРЬЯ'] == 10)
|( df['НОВЫЙ ТИП СЫРЬЯ'] == 11)
|( df['НОВЫЙ ТИП СЫРЬЯ'] == '?'), 'другое', df['НОВЫЙ ТИП СЫРЬЯ'])

df['НОВЫЙ ТИП СЫРЬЯ'] = np.where((df['НОВЫЙ ТИП СЫРЬЯ'] == 'серый, пятно')
|( df['НОВЫЙ ТИП СЫРЬЯ'] == '10А')
|( df['НОВЫЙ ТИП СЫРЬЯ'] == '10а'), '10A', df['НОВЫЙ ТИП СЫРЬЯ'])

df['НОВЫЙ ТИП СЫРЬЯ'] = np.where((df['НОВЫЙ ТИП СЫРЬЯ'] == 'серый, пятно')
|( df['НОВЫЙ ТИП СЫРЬЯ'] == '5А'), '5A', df['НОВЫЙ ТИП СЫРЬЯ'])

df['НОВЫЙ ТИП СЫРЬЯ'] = np.where((df['НОВЫЙ ТИП СЫРЬЯ'] == 'серый, пятно')
|( df['НОВЫЙ ТИП СЫРЬЯ'] == '2А'), '2A', df['НОВЫЙ ТИП СЫРЬЯ'])

df['НОВЫЙ ТИП СЫРЬЯ'] = np.where((df['НОВЫЙ ТИП СЫРЬЯ'] == 'серый, пятно')
|( df['НОВЫЙ ТИП СЫРЬЯ'] == '1А'), '1A', df['НОВЫЙ ТИП СЫРЬЯ'])
```

```python
grouped = (df.groupby('НОВЫЙ ТИП СЫРЬЯ').size().reset_index(name='Counts').sort_values('Counts'))
plt.figure(figsize=(8, 5))
grouped.plot.barh(x='НОВЫЙ ТИП СЫРЬЯ', y='Counts', legend=False)
plt.title('Распределение количества объектов по атрибуту "Новый тип сырья"')
plt.xlabel('Количество записей в табличке')
plt.show()
#ХЗ че такое тип сырья, просто причесал данные
```
