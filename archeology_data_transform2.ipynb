```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

path = '/content/output.csv'
df = pd.read_csv(path)
```

```python
df['ТЕРМИЧЕСКОЕ ВОЗДЕЙСТВИЕ'] = np.where((df['ТЕРМИЧЕСКОЕ ВОЗДЕЙСТВИЕ'] == 'серый, пятно')
|( df['ТЕРМИЧЕСКОЕ ВОЗДЕЙСТВИЕ'] == '0')
|( df['ТЕРМИЧЕСКОЕ ВОЗДЕЙСТВИЕ'] == 'да+')
|( df['ТЕРМИЧЕСКОЕ ВОЗДЕЙСТВИЕ'] == 'средний')
|( df['ТЕРМИЧЕСКОЕ ВОЗДЕЙСТВИЕ'] == 'дп')
|( df['ТЕРМИЧЕСКОЕ ВОЗДЕЙСТВИЕ'] == '10А- дп')
|( df['ТЕРМИЧЕСКОЕ ВОЗДЕЙСТВИЕ'] == '+'), 'да', df['ТЕРМИЧЕСКОЕ ВОЗДЕЙСТВИЕ'])
```

```python
n = 10

grouped = (
    df.groupby('КАТЕГОРИЯ')
    .size()
    .reset_index(name='Counts')
    .sort_values('Counts', ascending=False)
    .query('Counts > @n'))


plt.figure(figsize=(8, 5))
grouped.plot.barh(x='КАТЕГОРИЯ', y='Counts', legend=False)
plt.title('Распределение количества объектов по атрибуту "Новый тип сырья"')
plt.xlabel('Количество записей в табличке')
plt.show()
```

```python
df['КАТЕГОРИЯ'] = np.where((df['КАТЕГОРИЯ'] == 'отщепа фрагмент медиальный')
|(df['КАТЕГОРИЯ'] == 'микропластинка')
|(df['КАТЕГОРИЯ'] == 'пластинки фрагмент дистальный')
|(df['КАТЕГОРИЯ'] == 'отщеп первичный')
|(df['КАТЕГОРИЯ'] == 'пластинки фрагмент медиальный')
|(df['КАТЕГОРИЯ'] == 'пластинки фрагмент проксимальный')
|( df['КАТЕГОРИЯ'] == 'отщепа фрагмент проксимальный')
|( df['КАТЕГОРИЯ'] == 'пластина')
|( df['КАТЕГОРИЯ'] == 'отщепа фрагмент дистальный')
|( df['КАТЕГОРИЯ'] == 'пластинчатый отщеп')
|( df['КАТЕГОРИЯ'] == 'осколок')
|( df['КАТЕГОРИЯ'] == 'пластинка')
|( df['КАТЕГОРИЯ'] == 'чешуйка')
|( df['КАТЕГОРИЯ'] == 'отщеп'), df['КАТЕГОРИЯ'], 'другое')
```

```python
grouped = (df.groupby('КАТЕГОРИЯ').size().reset_index(name='Counts').sort_values('Counts'))
plt.figure(figsize=(8, 5))
grouped.plot.barh(x='КАТЕГОРИЯ', y='Counts', legend=False)
plt.title('Распределение количества объектов по атрибуту "Категория"')
plt.xlabel('Количество записей в табличке')
plt.show()
```
